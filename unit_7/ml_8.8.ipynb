```python
import numpy as np, pandas as pd
import matplotlib.pyplot as plt
from scipy import interpolate
from sklearn.metrics import mean_squared_error
```

```python
# plt.style.use('ggplot')

w1 = 10
a1 = 1
Fs = 30
Ts = 1.0 / Fs
alpha = 5

##   ( )
t = np.linspace(0, 1, Fs)
##   
y = a1 * np.exp(-alpha * t) * np.sin(2 * np.pi * w1 * t)


##   ( )
tt = np.linspace(0, 1, 100 * Fs)
##    (   )
y_high_res = a1 * np.exp(-alpha * tt) * np.sin(2 * np.pi * w1 * tt)

f, ax = plt.subplots(figsize=(12, 8))
ax.plot(t, y)
ax.plot(tt, y_high_res, "--")
ax.legend(["  ", "  "])
ax.grid()

plt.show()
```

```python
# interpolate.interp1d(tt, y)
y_interp = pd.DataFrame()
y_interp["t"] = np.linspace(0, 1, 100 * Fs)
y_interp["y"] = interpolate.interp1d(t, y, "cubic")(y_interp["t"])
# y_interp["y"] = np.interp1d(ti, df.t, df.y)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(y_interp["t"], y_interp["y"])
ax.plot(tt, y_high_res, "--")
ax.legend(["  ", "  "])
ax.grid()

plt.show()
```

```python
round(
    mean_squared_error(y_high_res, y_interp["y"]),
    5,
)
```

```python
w1 = 5
a1 = 1
Fs = 1000
Ts = 1.0 / Fs
alpha = 5

df = pd.DataFrame()
## 
df["t"] = np.linspace(-1, 1, Fs)
##   
df["y"] = np.sin(2 * np.pi * w1 * df.t) / df.t

##  
df["y_lost"] = df.y
df.loc[0::3, "y_lost"] = np.nan


f, ax = plt.subplots(figsize=(12, 8))
ax.plot(df.t, df.y_lost)
ax.set_ylim([-10, 10])
ax.legend(["  "])
ax.grid()
plt.show()

df.head(10)
```

```python
df["y_lost_nearest"] = df["y_lost"].interpolate(method="nearest")
df["y_lost_nearest"].fillna(method='bfill', inplace=True)
df["y_lost_nearest"].fillna(method='ffill', inplace=True)
round(
    mean_squared_error(df["y"], df["y_lost_nearest"]),
    3,
)
```

```python
df["y_lost_poly5"] = df["y_lost"].interpolate(method="polynomial", order=5)
# df["y_lost_nearest"].fillna(method='bfill', inplace=True)
# df["y_lost_nearest"].fillna(method='ffill', inplace=True)
df["y_lost_poly5"].fillna(0, inplace=True)
# df["y_lost_poly5"].fillna(method='bfill', inplace=True)
# df["y_lost_poly5"].fillna(method='ffill', inplace=True)
round(
    mean_squared_error(df["y"], df["y_lost_poly5"]),
    5,
)
```
