```python
import numpy as np
import matplotlib.pyplot as plt
import random
import statsmodels.api as sm
import matplotlib.pyplot as plt
plt.style.use('ggplot')

%matplotlib inline

CNT = 100
```

```python
eps = np.random.normal(size=CNT)
```

```python
x = list(eps[:2])
[x.append(0.2 * x[-1] + 0.1 * x[-2] + eps[i]) for i in range(2, CNT)]
plt.plot(x)
```

```python
test = sm.tsa.adfuller(x)
print("p-value: ", test[1])
if test[1] < 0.05:
    print(" ряд стационарен")
else:
    print("ряд не стационарен")
```
